### 2. Splitting Partitions
 
In this you must be concerned by losing data when partitioning your disk.

Types of partition scheme -  
1. MBR (master boot record)
2. APB (apple partitioning map scheme)
3. GPT (Guide partitioning scheme)

### Partition your disk

### 2a Destructive partitioning

#### Method1 (destructive) - Partition disk

To know how to use it - `diskutil PartitionDisk`

In order to partition, follow the following syntax to split that disk into two partitions is   
```
diskutil partitionDisk 
path_of_disk_operating_on (/dev/disk2)
no_of_partition (2)
Type_of_paritionScheme (MBR)
format_for_partition(atleast the format for the first partition)  (fat32)
name_of_that_first_partition (MBR1)
size(can be in kB or %) 
format for the 2nd partition (fat32)
name_for_second_partition (MBR2)
size (R) R stands for used up the rest of the space.
```
 
The code will look like the following. 
`diskutil partitionDisk /dev/disk2 2 MBR fat32 MBR1 50% fat32 MBR2 R`

The above command will partition your disk (I am using external USB) into two: MBR1 and MBR2

Now if you check the volumes - `ls -lG /volumes/`. The result is shown below
```
total 32
drwxrwxrwx@ 1 rajanbawa  staff  8192 Nov 16 04:09 MBR1
drwxrwxrwx@ 1 rajanbawa  staff  8192 Nov 16 04:09 MBR2
lrwxr-xr-x  1 root       wheel     1 Oct 28 01:56 Macintosh HD -> /
```

**Now this process is destructive as any important information that was on that drive would be lost during the partition.**

#### Method2 (also destructive) - split partition

`diskutil splitPartition` will give you information as to how to use it.

```
diskutil splitPartition MountPoint|DiskIdentifier|DeviceNode
        [numberOfPartitions] [(<fileSystemPersonality> <volumeName> <size>)*]
``` 

The actual code will be something like:- `diskutil splitPartition /dev/disk2s2 2 JHFS+ APM1 8G JHFS+ APM2 7.4G`

The explanation for the code is below-
```
/dev/disk2s2 (path of the disk that you want to partition)

2 (no. of partition)

JHFS+ (format for the first parition)

APM1 (name of the partition)

8G (size - you cannot use the %)

JHFS+ (format for the second partition)

APM2  (name of the second partition)

7.4G (size of the second partition)
```

#### Method3 (**Safest way to partition**) - resizeVolume

It is Non-destructive method. resizeVolume is only supported on a Journaled HFS+ file system.

To get the information on how to use it - `diskutil resizeVolume`

The best part - `diskutil resizeVolume bootvolume limits`. This will tell you the current size of your boot volume, how small you can make it

bootvolume - /Volumes/Macintosh\ HD 

To size the disk `diskutil resizeVolume disk_identifier 100G JHFS+ New 399G`

Understanding above code?  
the present disk will be divided into two.  
- existing disk will be resize to 100G 
- new disk will be resized to 399G and will be called New and will have JHFS+ file system.

__Note__ **resizeVolume is only supported on JHFS + file system. So you can't use it for APFS system. **

Benefit of using it- It doesnot destroy existing data on that disk.

## 6. What is a file system

It is a format in which files are stored in an operating system. It provides data structure (file hierarchy - main directory --> sub directory --> file1, file2, file3). It decides file permissions (read,write and execute). It allows you to track file location and retrieve file from your disk.

Also, Different OS are compatible with different file system, which means Ex FAT file system formatted extrnal hard-drive may not work on OS. APFS formatted external hard-drive may not work on earlier versions of Mac and it also is not workable file system for Windows.
    
## 7. Different types of file system   

| APFS | J HFS+ | Ex FAT |
|-------|-----|------|
| It stands for apple file system | It is also called Mac OS Extended system | It is produced by Microsoft |
| -It is apple's current default file system since 2016’s macOS Sierra | - It had been used as apple's default file system before apple released macOS Sierra | Ex FAT file format is used when working with both Windows and Mac. |
| -meaning any Mac using an older operating system will not be able to write to APFS-formatted drives. | - meaning if you have to work with Mac and you are using an external drive then J HFS+ format will probably be the safest option |  provide cross-platform compatibility  |
| - Benefit | | -Benefit |
| It is faster in terms of copying and pasting a folder. | An external drive used for Mac file backup can use JHFS + format | Formatting a drive with ExFAT offers one huge advantage: both Windows and macOS computers and both read and write to this format.   |
| it’s very quick to do things like determine how much space a folder is taking up on your drive. |   | Drawback -  It’s not a particularly optimized file format—it’s far more vulnerable to file fragmentation than APFS or Mac OS Extended, for one thing, and metadata and other features used by macOS aren’t present. | 
| - Security | |  |
| Both APFS and Mac OS Extended offer an Encrypted option | |  |
| problem 1- forgetting the encryption key means losing access to your files. | |  |
| problem 2 -  Reading and writing will be slower on an encrypted drive, but we think it’s generally worth it—especially on portable Macs, like laptops. | | |


## 8. Disk Partition 

Creating divisions within the same disk. 

Information about different volumes, their location and size can be known using `diskutil list` command.

Each partition has its unique identifier. This allows each partition to be a distinct "logical" disk. 

| Benefit | Downsides |
|----|-----|
| Each partition can use different file format system. For instance, disk2s1 may use FAT32 file system, whereas disk2s2 may use JHFS + file system | It can be difficult to properly size partitions, resulting in having one partition with too much free space and another nearly totally allocated
| This allows you to store different kind of file on the same hard-drive | |
| Adds Safety | |
| - For instance, One partition can just be used for storing system Data. Separating user data from system data can prevent the system partition from becoming full and rendering the system unusable. | |
| - Another partition can serve the use of backing up essential data. | |

The disk stores the information about the partitions' locations and sizes in an area known as the partition table that the operating system reads before any other part of the disk. 


## 9. Boot - (putting files into the brain of the computer)

You first download an operating system onto your computer or may have than OS on an external drive.

At this point, OS cannot be used by your computer. 

To use it, you must 'load' it on to your computer memory (RAM) which allows your CPU to read the files.



## 10. Different partitioning Schemes

| GPT  |  MBR |
|----|----|
| 


## 11. Merging theParitions

`diskutil mergePartitions` will provide you usage information. 

Let's try merging partitions - I have an external drive that has been previously partitioned using `diskutil PartitionDisk` in FAT 32 file format. 

Now I want to merge those partitioned disks into one and also I want to change the file format system to J HFS+.

Let's see how to do it

1. `diskutil mergePartitions file_format Name_of_merge_disk diskIdentifier_that_you_are_merging

```python
diskutil mergePartitions JHFS+ New disk2s1 disk2s2

Name of merge disk is New
Disks that I am merging- disk2s1 and disk2s2
```

When you press enter, you will a warning asking your permission
```
The chosen disk does not support resizing.
Do you wish to format instead? (y/N) 


Disk doesnot support resizing means it will be a data destructive process, which means everything on that disk will be destroyed. 
```

Note - If you have more than two partitions and you want to merge them all, then mention the first and the last identifiers. This will merge all disk.

How does this merging happens?  
It deletes the second partition and then grows the first one in a non-destructive manner. 

It means that any data stored in disk2s1 will not be destroyed, but it also means all the data on disk2s2 will be destroyed.
