# Violinplots vs Boxplots

## Violinplot( )

It is a combination of boxplot and kernel density estimate.

kernel density estimate tells you where the data points are most concentrated, while boxplot gives you the min, 25%,50%(median),75% and max info.

![](violinplot.png)

Action Genre was more width around the lower quartile, this means more data points are present within lower quartile (25%) and the median (50%).

This visual telling as to where more data points exist makes violinplot more appealing than boxplot.

### When to use violinplot( )
- When you have one numerical column and one categorical column. It is your choice as to what display you want, but keep categorical variable (e.g. Genre) on x-axis and keep numeric variable (e.g. BudgetMillion) on y-axis.

e.g. Genre vs Budget, Year vs CrticRating.

__Note-__ Don't use boxplot() or violinplot when both axes are numerics.

- When your sample size is large.

```python
z= sns.violinplot(data=movies,x='Genre',y='CriticRating') 

#you are also create for a filtered data
z1= sns.violinplot(data=movies[movies.Genre=='Drama'],x='Year',y='CriticRating')
```

### Benefit of using violinplot( )
- First is the visuals are better. kde telling as to where more data points exist makes violinplot more appealing than boxplot.

- Second, they help you understand boxplot better.

- Third, violinplot are more insightful compared to boxplot.

If you look at 2011 violin plot .. you see that there are two areas where the concentration of points is more.  
The first is at the top (major concentration), and second is at the bottom (minor concentration)

This is something that boxplot doesnot tell.


## Boxplot( )
### What do you see in a boxplot?  
- min,25%,50%(median),75% and max and outliers.

1. Median = Average

Average tells you as to where most data points exist. 

2. Boxplot is either compact or long.

Compact boxplot means volatility is less (e.g. When you draw CriticRating vs Genre, compact boxplot will mean that criticRating is not fluctuating too much)

Long boxplot means volatility is high, meaning fluctations are quite a lot .

3. The median is close to the lower quartile 

This means more concentration of data points is seen in the lower side of the median than you see above it. **This is more evident in violinplot**

4. The median is close to upper quartile

This means more concentration of data points is seen in the upper side of the median than you see below it.

````python
#Revision

import pandas as pd
import os

from matplotlib import pyplot as plt
import seaborn as sns

%matplotlib inline

import numpy as np

import warnings
warnings.filterwarnings('ignore')

sns.set_style('darkgrid')

#import data
os.getcwd()

movies = pd.read_csv('movies.csv')

movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating',
       'BudgetMillions', 'Year']

movies.Film=movies.Film.astype('category')

movies.Year = movies.Year.astype('category')

movies.Genre = movies.Genre.astype('category')

movies.Genre.cat.categories
movies.Genre.unique()

#Filtering
movies[2:10] #for rows use slicing
movies[['Film','Year']].head() #for columns, use double list. It is similar to [c('Film','Year')]

movies.Genre #This gives you specific column

movies[movies.Genre=='Action'].BudgetMillions #gives you BudgetMillions for Action Genre


#plots we covered
k1 = sns.lmplot(data=movies, x='AudienceRating', y='BudgetMillions',fit_reg=False,hue='Genre')

k2 = sns.jointplot(data=movies, x='AudienceRating', y='BudgetMillions',kind='hex',xlim=(20,80))

k3 = sns.kdeplot(data=movies, x='AudienceRating', y='BudgetMillions',shade_lowest=False,shade=True,cmaps='Reds')


````

````python
f,axes = plt.subplots(2,3,figsize=(12,6))
k1 = sns.kdeplot(data=movies, x='CriticRating', y='BudgetMillions',shade_lowest=False,cmap='Greens',shade=True,ax=axes[0,0])
k2 = sns.kdeplot(data=movies, x='AudienceRating', y='BudgetMillions',shade_lowest=False,camp='Greys',shade=True,ax=axes[0,1])
````

````python
f,axes = plt.subplots(1,3,figsize=(12,6))
k1 = sns.kdeplot(movies.CriticRating, movies.BudgetMillions,ax=axes[0])
k2 = sns.kdeplot(movies.AudienceRating, movies.BudgetMillions,ax=axes[1])
````
