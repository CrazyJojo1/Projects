# What is a JointPlot?

Draw a plot of two variables with bivariate graph (like scatter plot, hex plot, regression plot, etc.) and univariate graphs (like distribution plot for each of the two variables)

```python
from matplotlib import pyplot as plt
import seaborn as sns
#creating joint plot
plot = sns.jointplot(data=movies, x="AudienceRating",y="CriticRating")
```
![](Jointplot.png)

## use kind parameter

This allows you to change the bivariate graph

```
kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
    Kind of plot to draw. See the examples for references to the underlying functions.
```
### What is hex plot?
hex plot group cluster of dots. If you look at scatter plot, some dots are more concentrated in some areas. 

Area that contains more number of dots will form dark hexagons and area that contains less number of dots form lighter hexagons.
![](hex.png)

````python
import matplotlib.pyplot as plt

#you can also write the above as below

from matplotlib import pyplot as plt
````

````python
import seaborn as sns
%matplotlib inline  
#this print the plot in the same notebook we are working in 

import warnings
warnings.filterwarnings('ignore')
````

````python
#Revision and importing data
import pandas as pd
import os

movies = pd.read_csv('movies.csv')
#chanigng column names

movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating',
       'BudgetMillions', 'Year']

#changing Genre, film and Year into categorical variable

movies.Genre = movies.Genre.astype('category')
movies.Film = movies.Film.astype('category')
movies.Year = movies.Year.astype('category')

movies.info()
````

````python
#JointPlots

j= sns.jointplot(data=movies,x='CriticRating', y='AudienceRating')

#What do you see?
#The joint plot contains a scatter plot in the middle and also a histogram plot for each variable
````

````python
# What do you see in the scatter plot

#Audience raing start at 20, whereas Critic rating starts at 0. This shows Audience is more forgiving than critic.

#What about histograms (distribution plots)
#- Audience Rating is more like Normal distribution.
#- Critic Rating is uniform distribution
````

````python
#additional parameters used in joint plot

#1. kind parameter
#Change scatter plot to something else

plot = sns.jointplot(data=movies,x='CriticRating', y='AudienceRating', kind='hex') #This gives you hexagon diagram 

#Benefit of using hex plot?

#it allows you to see where the cluster of data is. It group the dots. 
#Darker hexa means group of dots are concentrated in that area

#lighter hexa means less dots are there
````

````python
plot = sns.jointplot(data=movies,x='CriticRating', y='AudienceRating', kind='kde')
````
