Below are couple of things that we can change in this plot-

## 1. style of the grid

Even though we are using pyplot function ($plt.subplots()$), using seaborn setting will still add changes on top of your pyplot charts. 

To have whitegrid background - use `sns.set_style('whitegrid')` 

## 2. Change the size of the chart

**The hack is to create a subplot with only just one item**

```python
f, axes = plt.subplots() #not specifying any parameter like nrows and ncols

#now even though you have not specified the axes, the plot will be placed in this subplot

```

### Why are first creating a subplot and placing our very only plot inside it?
This provides you flexibility to easily adjust the size of this plot.

You can adjust your chart to an A4 size sheet as shown below. The size is adjusted in units of an inch.

```python
f.set_size_inches(11.7,8.27) #This is size of A4 paper
```

## 4. Adding labels to the x and y axis

```python

```

## 5. Change the font size of the ticks (ticks are numbers that you see on each of your axis)

```python

plt.xticks(fontsize=20)
```

## 6. Legend Characteristics

### 6a. Increasing the size of the legend

We have already used `plt.legend()` 

But now we just need to add additional parameters into this function 

### 6. modify the size using 'prop'

This is the name of dictionary that helps you adjust the size. `plt.legend(prop={'size':20})`

### 6b. Changing other characterisitcs of your legend

You might see a grid behind your legend if your style is "Whitegrid" or "darkgrid")

1. use frameon - it will put an plain background for legend
2. use fancybox - it will provide curved corners to legend
3. facecolor - provide the fill for the background of the legend
4. edgecolor - provides color to the borders of the legend

```python
plt.legend(prop={'size':20},frameon=True,fancybox=True, shadow = True, framealpha=1,facecolor='silver',edgecolor='Black')
plt.show()
```
## 7. Legend location and gap from the plot

### 7a. legend outside of the plot - use ``bbox_to_anchor``

```python
plt.legend(prop={'size':20},frameon = False, bbox_to_anchor=(1,1))

top right corner- bbox_to_anchor=(1,1)
bottom right corner- bbox_to_anchor=(1,0)
top left corner- bbox_to_anchor=(0,1)
bottom left corner- bbox_to_anchor=(0,0)
```
### 7b. legend inside of the plot - use ``loc``

```python
plt.legend(prop={'size':20},frameon = False,loc=1)
#loc=1 is for top right corner
#loc=2 is for top left corner
#loc=3 is for bottom left corner
#loc=4 is for bottom right corner
```

### 7c. To increase the gap from the borders of the plot - use `borderaxespad`

```python
plt.legend(bbox_to_anchor=(1, 1), borderaxespad=5)

#borderaxespad is used to increase the gap from the borders from the top and from the right side
```
