# Histogram

## Normal Distribution vs Uniform Distribution
| Normal Distribution | Uniform Distribution |
|:-----------------:|:----------------:|
| an arrangement of a data set in which most values cluster in the middle of the range and the rest taper off symmetrically toward either extreme | Type of probability distribution in which all outcomes are equally likely |
| Seen where emotions play an important part (normal human behavior) | Seen where emotions are not playing an important part |
| As the sample size increases, most human behavior fall in normal distribution | Where rules and objective analysis is used, you see uniform distribution. e.g. Critic doing movie rating based on sound effects, lighting, plot and other rules |

## What functions can be used to create histogram plots

`sns.distplot(column)` 
`plt.hist(column)`

```python
plot1 = sns.distplot(movies.AudienceRating)
plot2= plt.hist(movies.CriticRating)
```

### changing styling

