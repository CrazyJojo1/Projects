# Building Dashboards in python

## What is dashboard?

It is a combination of charts. A dashboards is a way to present different plots side by side. It is pretty much what we do when we use subplots function.

To create a dashboard we will be using $subplots()$

___

### Creating dashboard using seaborn functions - `kdeplot( )` `violinplot( )` `boxplot( )`, etc

```python
#creating subplot
f,axes = plt.subplots(nrows=2,ncols=2,figsize=(15,15))
#This means plt.subplots return two values: f (figure) and axes (marks the location for each figure)

#let's start populating these figures
plot1 = sns.kdeplot(x=movies.BudgetMillions,y=movies.AudienceRating,cmap='Reds',shape=True,shade_lowest=False,ax=axes[0,0])
plot2 = sns.kdeplot(x=movies.BudgetMillions,y=movies.CriticRating,cmap='Reds',shape=True,shade_lowest=False,ax=axes[0,1])
plot1.set(xlim=(-20,160))
plot2.set(xlim=(-20,160))
plot3 = sns.violinplot(data = movies[movies.Genre=='Drama'],x='Year',y='CriticRating',ax=axes[1,0])
plot4 = sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='Reds',shade=True,shade_lowest=False,ax=axes[1,1])
plot4.outline =  sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='Reds')

plt.show()
```

## Using $subplots()$ with pyplot charts is different than using it for seaborn charts.

### What you need to know about axes?

__axes is a pyplot object__
```
f,axes = plt.subplots() means f(figure) and axes (set of objects) are two objects generated by pyplot.

An object generated by pyplot can easily access the functions in-built in pyplot library (The likes of  `scatter( )` `hist( )` `plot( )`)

This is why you specify the object (axes) first and then the function when using pyplot function.
```
```python
f,axes = plt.subplots(nrows=2,ncols=2,figsize=(15,15))
axes[1,1].hist(movies.CriticRating,bins=30,rwidth=1) 
```

Seaborn on the other hand is add-on on top of pyplot.

```
This means, axes (the object of plt) will have no knowledge of things contained within seaborn packages. As a result, axes has no visibility of .kdeplot(), violinplot() and so on.

so you cannot call axes[1,1].violinplot() 
```
```python
#You will get an error message if you do the following
ax=axes[1,0].violinplot(data = movies[movies.Genre=='Drama'],x='Year',y='CriticRating')
```

**Conclusion-**  

You can call `axes[1,1].scatter()` (axes, the object and scatter, the function, are both part of pyplot package).  
This allows axes to use scatter function.

But same is not true for violinplot.    
This is why seaborn function is specified first and you add axes as a parameter to it.

pyplot is a base package whereas seaborn is an an add-on to pyplot. 

## For more information

[learn about pyplot](https://matplotlib.org/2.0.2/examples/pylab_examples/subplots_demo.html)

````python
#creating subplot
f,axes = plt.subplots(nrows=2,ncols=2,figsize=(15,15))
#This means plt.subplots return two values: f (figure) and axes (marks the location for each figure)

#let's start populating these figures
plot1 = sns.kdeplot(x=movies.BudgetMillions,y=movies.AudienceRating,cmap='Reds',shape=True,shade_lowest=False,ax=axes[0,0])
plot2 = sns.kdeplot(x=movies.BudgetMillions,y=movies.CriticRating,cmap='Reds',shape=True,shade_lowest=False,ax=axes[0,1])
plot1.set(xlim=(-20,160))
plot2.set(xlim=(-20,160))
plot3 = sns.violinplot(data = movies[movies.Genre=='Drama'],x='Year',y='CriticRating',ax=axes[1,0])
plot4 = sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='Reds',shade=True,shade_lowest=False,ax=axes[1,1])
plot4.outline =  sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='Reds')

plt.show()
````

````python
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
````
