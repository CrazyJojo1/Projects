__How to create a function:__ 
___

__SYNTAX__
```python
def function_name(parameters):
    body of the function(code)
```    

Let's create our first function

```python
def myplot():
    plt.plot(Accuracy, c='black', ls ='dotted', marker='s', ms = 7)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.show()
#to run this funciton 
myplot() 
#The output will have accuracy plot for all ten players
#-------------------------------------------------------------
#now we need to use this function to plot specific player
#goal:
myplot(["KobeBryant","JoeJohnson","LeBronJames","CarmeloAnthony"])
#in order to achieve this, we will need to provide our functioon with some parameters
def myplot(a,b):
    for i in range(a,b):
        plt.plot(Accuracy[i], c='black', ls ='dotted', marker='s', ms = 7,label=Players[i])
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.show()
#print the output
myplot(1,4)
```
**Problems with above function**

The limitation of above code- <br>
(a,b) are range parameters - (start,stop) <br>
(a,b) = (1,4)
for i in range(1,4) will give i three values = 1,2,3 <br>
so, <br>
For Loop will run three iterations <br>
player2,player3 and player4 will be plotted <br>
But, we cannot get palyers of our choice <br>

How to fix this issue  
**The Fix**
```python
def myplot(playerList):
    for string in playerList:
        plt.plot(Accuracy[Pdict[name]], c='black', ls ='dotted', marker='s', ms = 7,label=Players[Pdict[name]])
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.show()
#use the function
myplot(["KobeBryant","JoeJohnson","LeBronJames","CarmeloAnthony"])
```
**Limitation of above code-**

All players are plotted in same color and using same marker. 

This is making it difficult to tell who is who.

Sol??
**Changing the color and marker for each player**
- create dictionary for color using player's name as key
- create dictionary for marker using player's name as key 

Why player's name need to be the key?? 

Since we are already passing player name's as string in For Loop using playerList, it will come handy to evoke color and marker using the same string 
```python
Cdict = {"KobeBryant":'Blue',"JoeJohnson":'orange',"LeBronJames":'green',"CarmeloAnthony":'red',"DwightHoward":'purple',"ChrisBosh":'brown',"ChrisPaul":'pink',"KevinDurant":'gray',"DerrickRose":'olive',"DwayneWade":'cyan'}
Markerdict = {"KobeBryant":'o',"JoeJohnson":'v',"LeBronJames":'^',"CarmeloAnthony":'<',"DwightHoward":'>',"ChrisBosh":'s',"ChrisPaul":'p',"KevinDurant":'*',"DerrickRose":'h',"DwayneWade":'d'}
```

Let's use these in our code now  

```python
def myplot(playerList):
    for string in playerList: #playerlist is just like a number generator
        plt.plot(Accuracy[Pdict[string]], c=Cdict[string], ls ='dotted', marker=Markerdict[string], ms= 7,label=string)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.show()
```

**Limitation of above function:-**

It can only be used for single matrix.  

Sol??  

What if we can provide matrix as our parameter.

```python
def myplot(data, playerList):
```

Can we improve this function further:  
Yes.  
How?  
**Add default values.**  
Benefit?  
If playerList parameter is not declared when calling the function, then function will automatically assume the default values.
```python
def myplot(data, playerList = Players):

        plt.plot(data[Pdict[string]], c=Cdict[string], ls ='dotted', marker=Markerdict[string], ms= 7,label=string)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.show()
#use function 
myplot(Salary) #I am not specifying playerList and still no error.
```

**Adding title to your plot-** 
```python
def myplot(data, 'title', playerList = Players):
    for string in playerList: #playerlist is just like a number generator
        plt.plot(data[Pdict[string]], c=Cdict[string], ls ='dotted', marker=Markerdict[string], ms= 7,label=string)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.title(title)
    plt.show()
```    
**Few things you must know:-**
    
1. In order to add Legend to your plot, you must use 'label' in your code.

1. you need to specify the titles in quotation marks when calling your function 
```python
myplot(FieldGoals/FieldGoalAttempts, 'FieldGoals/FieldGoalAttempts')
#title is in quotation marks
```

1. Order of parameter is important: non-default parameters must come first. 
```python
def myplot(data, 'title', playerList = Players):
#since playerList has default values, you place it last
```
