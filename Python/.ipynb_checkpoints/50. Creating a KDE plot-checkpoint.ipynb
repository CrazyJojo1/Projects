## What is Kernal density estimate?
Kernel means the centre. 

Kernel density means centre where most data points are collected

Kernel density estimate gives you an ideas to where on average most data points exist. Points are more concentrated around the centre and as the distance form the centre increases, density keeps on decreasing.


## How to create a KDE plot? - use `sns.kdeplot( )`

```python
sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating)
```

### styling a KDE plot 

Use
- shade = True
- shade_lowest=False (This will get rid of dim boundaries and this will also give you the grid if you have `sns.set_style('darkgrid')`
- cmap = 'Reds' (This will change the shade color)
- You have darker outline, lay basic KDEplot over stypled plot as is shown below

```python
#This is styled kdeplot
plot1 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating, shade=True,shade_lowest=False,cmap='Reds')

#This is basic kdeplot
plot2 = sns.kdeplot(x=movies.CriticRating, y=movies.AudienceRating,cmap='Reds')

#Run both plots in the same cell.
#This will darken the outline and will give better visuals
```

````python
#Revision

#importing all essential packages

import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

import warnings 
warnings.filterwarnings('ignore')

%matplotlib inline
plt.rcParams['figure.figsize']= 8,4

#to import csv file
import pandas as pd 

#to check and change current working directory
import os 
#--------------------------------------

os.getcwd()
movies = pd.read_csv('movies.csv')

#exploring data
movies.head()
#changing column names
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating',
       'BudgetMillions', 'Year']


#creating factor or categorical variables

movies.Genre= movies.Genre.astype('category')
movies.Film= movies.Film.astype('category')
movies.Year= movies.Year.astype('category')

movies.info()
movies.Genre.cat.categories

movies.Year.cat.categories

#creating lmplot
sns.set_style('darkgrid')

plot1= sns.lmplot(data=movies, x="CriticRating", y="AudienceRating",hue='Genre',fit_reg=False,size=10)

````

````python
#creating jointplot
plot2 = sns.jointplot(data=movies,
                     kind='hex',
                     x='AudienceRating',
                     y='CriticRating'
                     )
````

````python
#lastly loops
#What needs to be iterated?
#i need to create a stack plot that contains a list. each list object tells Budget of that Genre.

#so I need to create a list that contains - movies[movies.Genre==''].BudgetMillions

MyList = []
Mylabel= []

for string in movies.Genre.cat.categories:
    MyList.append(movies[movies.Genre==string].BudgetMillions) 
    Mylabel.append(string)
    
print(Mylabel)

plot5 = plt.hist(MyList,bins=30,stacked=True,rwidth=1,label=Mylabel)
plt.legend()
plt.show()


````
