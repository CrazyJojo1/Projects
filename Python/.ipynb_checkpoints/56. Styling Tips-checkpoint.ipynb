## 1. Changing the background for seaborn functions

- To change the Background - `sns.set_style('dark',{"axes.facecolor":"black"})`

In order to change the background, you will need to create a dictionary in seaborn style function - `sns.set_style()`.

Now sns.set_style() function can have limited set of values ('white','whitegrid','dark','darkgrid')

use Dictionary in sns.set_style as shown 
```python
sns.set_style('dark',{"axes.facecolor":"black"})

or 

sns.set_style('dark',{"axes.facecolor":"Yellow"})
```

The Dictionary uses axes.facecolor, a keyword, to add a different background to your seaborn functins.

## 3. Styling a Kdeplot

### 3a. Outline - This will add definition to the borders.

In order to create an outline, you can superimpose a similar plot onto existing plot, but without shade parameter. To better understand, see below given code.

**Note- Use same axes for both the plots.** Notice, ax=axes[1,1] is same for both the plots. 

Also, adjust the same xlim and ylim for both the plots.

```python
#plot4 [1,1]
plot4 = sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='Blues_r',\
                    shade=True,shade_lowest=False,ax=axes[1,1])

plot4.set(xlim=(-50,250))

#outline for plot4 - 
outline4 =  sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='gist_gray_r',\
                             ax=axes[1,1])
outline4.set(xlim=(-50,250))          
```


### 3b. shade = True 

This is used to add fill to your plots. 

### 3c. shade_lowest = True 

This is used to give color to the background. (Background is anything than the main diagram)

### 3d. cmap parameter- It is used in conjunction with shade=True. 

cmap provide different color schemes for the fill ('shade')

This is very similar to palette, a parameter that we used in violinplot. 

Here are the recommendations for cmap color schemes-

| shade | outline |
|:-----:|:-------:|
| 1. 'inferno_r' | 'cool' |
| 2. 'Blues_r' | 'gist_gray_r' |


## 4.Styling the boxplot

### 4a. Remove outliers from the boxplot - use `showfliers=False`

```python
plot2 = sns.boxplot(data=Releases, x='Genre',y='GrossUSPercent',color = 'lightgray',showfliers=False)
```

### 4b. To add transparency - define `alpha` in `plt.setp( )`

You cannot define alpha as a parameter inside $sns.boxplot()$. So instead you define it using $plt.setp( )$

```python
plt.setp(ax.artists, alpha=0.5)
```

### 4c. To add jitters to the boxplot - use ``sns.stripplot( )``

```python
plot = sns.stripplot(data=Releases, x='Genre',y='GrossUSPercent',hue='Studio',jitter=True,size=10,alpha=0.7)

# parameters used for Stripplot( ):-
size - defines the size of dots
alpha - defines transparency
jitter - will add jitters 
```

````python
## Revision

#import packages

import pandas as pd
import os

import warnings 
warnings.filterwarnings('ignore')

from matplotlib import pyplot as plt
import seaborn as sns

sns.set_style('dark',{'axes.facecolor':'Black'})
%matplotlib inline

#import files

os.getcwd()

movies = pd.read_csv('movies.csv')

movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating',
       'BudgetMillions', 'Year']

movies.Film = movies.Film.astype('category')

movies.Genre = movies.Genre.astype('category')

movies.Year = movies.Year.astype('category')


#movies.info()
#creating subplot
f,axes = plt.subplots(nrows=2,ncols=2,figsize=(15,15))
#This means plt.subplots return two values: f (figure) and axes (marks the location for each figure)

#let's start populating these figures

#Plot1 [0,0]
s1= sns.kdeplot(x=movies.BudgetMillions,y=movies.AudienceRating,shade=True,shade_lowest=True,ax=axes[0,0])
s1.set(xlim=(-20,160))

#adding an outline

s2 = sns.kdeplot(x=movies.BudgetMillions,y=movies.AudienceRating, cmap='cool',ax=axes[0,0])
s2.set(xlim=(-20,160))
#plot2 [0,1]
plot2 = sns.kdeplot(x=movies.BudgetMillions,y=movies.CriticRating,cmap='Spectral',shade=True,shade_lowest=False,ax=axes[0,1])
plot2.set(xlim=(-20,160))

s3 = sns.kdeplot(x=movies.BudgetMillions,y=movies.CriticRating,cmap='summer',ax=axes[0,1])
s3.set(xlim=(-20,160))
#plot3 [1,0]
plot3 = sns.violinplot(data = movies ,x='Year',y='BudgetMillions',ax=axes[1,0],\
                      
palette = 'YlOrRd')

#plot4 [1,1]
plot4 = sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='Blues_r',shade=True,shade_lowest=False,ax=axes[1,1])

#outline for plot4
plot4.outline =  sns.kdeplot(x=movies.CriticRating,y=movies.AudienceRating,cmap='gist_gray_r',ax=axes[1,1])



plt.show()
````
