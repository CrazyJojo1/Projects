# What is Git?

It is a version control system that allows you to make incremental changes to your file (code file). And, It gives you the ability to see your old code (anytime anywhere)

- Git allows you to write a bunch of code.

- collaboration tool  
The code written by you can be looked at by others (if settings is public) and also other coders can modify that code (add little bit more to it) for good. 

- It is also a version control tool.
    + Gives you backup files
        - If you do make a mistake, you can compare your existing code with earlier version of the code to see what is wrong and fix your mistake.
    + Keep a history
        - Every modification to your code is saved in Git database
    + Experiment
        - You are use a specific section of code taken from one of Git repositories and experiment with it.
        
- it is a distributed system
The entire project, including all of its history, is mirrored on the person's computer who are collaborating. 

- it is an open source and is free
This means anyone can download the package from Git. And coders can also contribute to it.

## Why distributed system is faster?
- Speed 
    - Since you are not using any central server, it takes out the performance of the network out of the equation.
    
- offline access
     - All changes you make are saved in your file in your computer. And since file is saved on your computer, you don't need to connected to the network to work on it.

### What are commits?

Since Git is an open source that allows everybody to contribute. When one of the contributor writes a code and packages that code in their or someone-else repository on Github it is called a commit.

So you can say commit is 
  - **some little update to a large repository.**
  - way to travel through a project's history and see codes at different point in time
      - **click on (< >), allows you to browse the repo from that point in history.**
      This means that you can see what the code looked like at the time when that commit was submitted.
  - hash (commits unique identifier)
      - a unique hash is given to each commit.
      - if you click on the first seven letters (e.g. c165870), then it will take you to the commit
  
One repo can have unlimited number of commits. It contains-
- date the commit was submitted
- name of the person
- header telling what that commit is about
- SHA, a unique identifier telling who submitted it.

**There are two ways to view the commit - split and unified**

- Split - The code info is split in two section-  
  - left side (deleted and is in red)
  - right side (changes added and is in green)
- Unified  
As the name suggest, you have complete file on one page containing deleted (red) and added (green) on same page

### Configuring Git on your computer

In order to know your global username and global email-address..

Why do you need to do this?  
Suppose you want to submit a commit.  
You pack your code and submit it.   
How will Git know who that commit is from? 

To do this, type

for username - `git config --global user.name "Your user name"` (if you have already done it press ctrl + C)

for email-address - `git config --global user.email "email address that you use to sign-up on Github.com"`

To check username and email - `cat ~/.gitconfig` (~/. means home folder of your computer)

### Editing files on your computer

In order to open .md file

step 1: Go into the directory (folder) containing that file

step 2: once in it, type `open -a Rstudio README.md`, this will open README.md file in Rstudio.

step 3: start editing your file in Rstudio. And press command + S to save changes

step 4 : verifying if the file was modified or not - To do this type `git status` 

You will see something like modified:   README.md in Red color.

step 5 : In order to see the changes on your README.md file, type `git diff README.md` (`diff` is used to compare new with old file)

You will see `+This is new edit from your local computer` in Green (The new line added to README.md file)

step 6: **Staging**

Even though your file is modified, your old file is not updated yet. 

How do I know that?

When you type `git status`... you will see something like 
```
(Changes not staged for commit:
  (use "git add <file>..." to update what will be committed)
  (use "git restore <file>..." to discard changes in working directory)
```
 
To add changes to your file, type `git add README.md`  (remember - command is case-sensitive)

Now if you type `git status` - you will have README.md in Green color.

So what are we doing here? (__Staging__)

You have prepared (staged) your file to be sent to the git. __It is not submitted yet.__

In order to submit it, we need to mention who this is to and what is update about.

___

### Your first commit (Submitting the file using terminal)

- __Writing a message as to what commit is all about__

`git commit -m "Updated the readme file" ` (-m means message (message is in double quotation as it is a string)

Now if you type `git status` - you will see 

```
Your branch is ahead of 'origin/main' by 1 commit.
  (use "git push" to publish your local commits)
```  
- Now for more details - `git log --oneline`. You will see

```
c6c9f74 (HEAD -> main) updated the readme file
3c1cf09 (origin/main, origin/HEAD) Initial commit
```

**Your code is still not at GitHub.com yet**

___

### Pushing files to GitHub repository

- type `git push origin main`

__NOTE__ When you enter this command, you will be asked username and password 

Username is same as that you see on Git.

Password is personalised token (Before generating your personal token - delete your existing Git password saved on computer (go to `Key Chain` --> type git --> right click to `Delete`)

After deleting your saved Git passwords, follows these steps
___


![](Create_personal_access_token_on_GitHub.png)

___

Generate tokens are valid for set timeline. So do take note of that as well.

**Revision**
```python
cd test-dir/test-repository/
 1  git status  #this will tell you your current status
 2  git diff README.md  #This tells you the changes that you made
 3  git add README.md  #This is Staging (think of it as putting your letter in an envelope)
 4 git status
 5 git commit -m "Second update" #signing that envelope (telling why you are sending it)
 6 git log --oneline 
 7  git push origin main  #This will push it to Github.com (you may need your username and token password for it)
 8  git log --oneline
```

Now, when you will hit a refresh on your github page, it will update your README.md file.

### How to download files from your repos (present at Github.com)

- You can also edit the .md file on Github as well. 

- Click on the pencil icon on top of README.md file and start making changes

- When done, in the **Commit Changes** seciton - write what this update is all about  
(This is similar to `git commit -m "I am updating my file"`

- Lastly, select, commit directly to the main branch.

- Done!

Benefit of doing on gitHub is that you don't need to use add and push command (you still will need to write commit changes - message)

**Problem - The changes will not be on your computer (It will only exit on github.com)**

When You open your terminal and type `git log --oneline`  
The latest changes in your .md file will not exit (the one that you made directly on github)

**Solution**
You will need to __Pull__ these latest changes. (Earlier we did `git push origin main`, but this time we are pulling files from github to our local computer)


To do this type - `git pull origin main`

Now if you do `git log --oneline`

You will see
```python
1670b86 (HEAD -> main, origin/main, origin/HEAD) update the readme for a third time
051fb5d Second update
c6c9f74 updated the readme file
3c1cf09 Initial commit
```
Here origin stands for github, main is the branch from which you are pulling your file.
