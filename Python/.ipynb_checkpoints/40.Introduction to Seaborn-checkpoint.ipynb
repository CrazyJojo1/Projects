# Introduction to Seaborn (Very similar to ggplot package)

## Seaborn is a powerful visualisation package for python.

- It builds on top of matplotlib. This means, Seaborn will make plot obained using matplotlib better. It does that by providing extra funcitons that are not available in matplotlib.
    * This also means, you must have matplotlib imported in order to use Seaborn package.
    
```python
import matplotlib.pyplot as plt
import seaborn as sns
#..................................other useful things
#in order to print the plot in the same notebook we are working on 
%matplotlib inline
#ignore warnings
import warnings
warnings.filterwarnings('ignore')
#adjust the size of the plot
plt.rcParams['figure.figsize'] = 8,4
```
In order to get the better understanding of the types of plot you can get using Seaborn package. Click on the link below:  
[Seaborn Gallery](https://www.python-graph-gallery.com/seaborn/#google_vignette)

**Tip on how to use this link:** Click on any chart. It will take you the web-page that will contain every detail on how to use it.

### Benefits:
- It allows you to draw insights charts (Boxplot, scatterplot, density charts, histograms) 
- color and visuals are appealing as well.

## How to use seaborn 

**Histogram using $.distplot()$**
```python
vis1 = sns.distplot(stats["InternetUsers"], bins=30)
```

**Boxplot using $.boxplot( )$**
```python
vis2 = sns.boxplot(data=stats, x="IncomeGroup",y="BirthRate")
```
**Density distribution using $.kdeplot( )$**
```python
plot2 = sns.kdeplot(stats["InternetUsers"])
plot2.set_title('A distribution chart made with Seaborn')
```

**Linear regression plot using $.lmplot( )$**
```python
vis3 = sns.lmplot(data=stats, x="InternetUsers", y='BirthRate',fit_reg=False, hue="IncomeGroup",size=5,aspect = 3)
#parameters used
1. fit_reg : removes the correlation line
2. hue : it is usd to add color to your plot (Using hue="columnname" is called mapping)
3. size : this parameter determines the size of your plot (lmplot size is 'not' controlled using plt.rcParams['figure,figsize']  
4. aspect = width/height (aspect = 1 (meaning width of plot is equal to its height), aspect = 2 (width is twice the height)
```
