````python
#Now let's add some colors and line styles

#parameter
#1. c = color
#2. ls = line style(whether you want it dotted)
#3. marker = how you want the dots to be represented (square dots, round dots, triangle dots and so on..)
#When you are plotting each player salary on a plot, marker can allow you to tell which player line you are looking at.
#4. ms = marker size e.g. ms=7
plt.plot(Salary[0], c = 'black', ls ='-.', marker = 's',ms=7) #marker = 's' means square shape 

#You can use help window (Matplotlib reference) 
#in its search window type 'plot linestyles'   this will allow you to see what other linestyles are there
````

````python
#You are seeing this line in the ouput of your plot
# [<matplotlib.lines.Line2D at 0x7ff76383eac0>]

#how to get rid of it
plt.plot(Salary[0],c='green',ls = '--',marker= 's',ms=7)
plt.show()  #plt.show is just like print()
````

````python
#Inorder to increase the size of your plot so that it expands to the right
plt.rcParams['figure.figsize']=8,4

#Here we are setting the parameters of our session - figure.figuresize (8= width , 4= height)
#play around with it to see different sizes
````

````python
#Now run your code again

plt.plot(Salary[0],c='green',ls = '--',marker= 's',ms=7)
plt.show() 

#looks like your plot has expanded a little to the right.
````

````python
#Now let's work on x-axis label and y-axis label

#plt.xticks(what is their on x-axis(we have list(range(10))), name you want on x-axis)


plt.plot(Salary[0],c='green',ls = '--',marker= 's',ms=7)
plt.xticks(list(range(0,10)), Seasons)  #Season is a list that we have created in our imported data
plt.show() 
````

#### Takeaways: <br>
1. First let's import numpy: this is used for statistical data```import numpy as np```.  <br>
2. Import pyplot of matplotlib : ```import matplotlib.pyplot as plt```   <br>
Note: matplotlib already is already installed in Anaconda. So, you just need to import it. <br>
Also ```plt``` is an alias for plyplot, which means you can use plt instead of plyplot when calling it. <br>

<u>How to use pyplot<u/> - <br>
    
    
```plt.plot(array_name or matrix name)``` <br>
```python
Salary[0] = array([15946875, 17718750, 19490625, 21262500, 23034375, 24806250,
       25244493, 27849149, 30453805, 23500000])
#use plyplot to plot Salary[0] 
plt.plot(Salary[0])
```
<u> How to get the plot in Jupyter notebook rather than putting them in separate window <u/> <br>
- use ```%matplotlib inline``` <br>

<u>Adding aesthetics to your plot <u/> <br>
    
Now let's add some colors and line styles <br>

1. c = color <br>
2. ls = line style(whether you want it dotted) e.g. ls='dotted' or ls='..' (use representation or word) <br>
3. marker = how you want the dots to be represented (square dots, round dots, triangle dots and so on..) <br>
Note: When you are plotting each player salary on a plot, marker can allow you to tell which player line you are looking at. <br>
4. ms = marker size e.g. ms=7 <br>
```plt.plot(Salary[0], c = 'black', ls ='--', marker = 's',ms=7)``` #marker = 's' means square shape  <br>


<u>use help window for Matplotlib reference<u/>:-  <br>
    
-In order to take help- <br>
a. click on Matplotlib reference in the help window <br>
b. in its search window type 'plot linestyles' . This will allow you to see what other linestyles you can use. <br>

<u>How to get rid of  ```[<matplotlib.lines.Line2D at 0x7ff76383eac0>]``` that you are seeing in the ouput of your plot<u/>  <br>
- use ```plt.show()``` <br>

<u>increase the size of your plot- <u/>   let's say you want your plot to expand to the right <br>
    
```plt.rcParams['figure.figsize']=8,4``` <br>
    
What are you doing here? <br>
    - Here we are setting the parameters of our session - figure.figuresize (8= width , 4= height) <br>
    
<u>How to change what you see on x-axis and its orientation <u/> <br>
- use ```plt.xticks()``` <br>
Syntax :- plt.xticks(what is their on x-axis(we have list(range(10))), name you want on x-axis) <br>
```python
plt.plot(Salary[0],c='green',ls = '--',marker= 's',ms=7)
plt.xticks(list(range(0,10)), Seasons)  #Season is a list that we have created in our imported data
plt.show()
```
- let's change the orientation of x-labels<br>
To do that- use paramete```rotation='vertical'``` <br>
```python
plt.plot(Salary[0],c='green',ls = '--',marker= 's',ms=7)
plt.xticks(list(range(0,10)), Seasons, rotation = 'vertical')  #Season is a list that we have created in our imported data
plt.show()
```
