# Takeaway

* __How to print name and the output in two different lines__

In order to print the name in one line and the output in the next line. Use `'\n'`

```python
#creating our first matrix
FreeThrows = np.array([KobeBryant_FT,JoeJohnson_FT,LeBronJames_FT,CarmeloAnthony_FT,DwightHoward_FT,ChrisBosh_FT,ChrisPaul_FT,
                      KevinDurant_FT,DerrickRose_FT,DwayneWade_FT])
print("FreeThrows: "'\n', FreeThrows)
```

Note: you don't need to use commas before `'\n'` as is shown in `print("FreeThrows: "'\n', FreeThrows)`.

+ __How to delete vectors (lists) that you don't need anymore__ 

Use `del()`

```python
# Remove vectors - we don't need them anymore
```
