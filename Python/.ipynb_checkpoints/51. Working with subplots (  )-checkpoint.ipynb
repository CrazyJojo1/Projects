# Working with subplots( )

## Benefit of using subplots( ) function?

Allows you to create more sophisticated plots in python by combining visualisation together.

What does this mean?, to simply put, it allows you to create multiple plots (subplots) side by side.

## The syntax of subplots( )
f,axes = plt.subplots(rows, columns, figsize = (width, height))
plot1 = (-----------------------,ax=axes[0])
plot2 = (------------------------,ax=axes[1])

```python
f,axes = plt.subplots(1,2,figsize=(12,6))
k1 =sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,ax=axes[0])
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,ax=axes[1])
```

### Understanding the code
f,ax = plt.subplots(1,2) will create a single row and 2 columns in which you can place your plots.  
This way you create your empty plots.

-----------------------------
f,ax=f,axes = plt.subplots(1,2,figsize=(12,6)) 

Here, you are adjusting the size of your empty plots.

-----------------------------

f,axes = plt.subplots(1,2,figsize=(12,6))  
k1 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,ax=axes[0])  
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,ax=axes[1]) 

Here, you are placing the plots in your empty spaces.

**Note**- It is important to use `ax=axes[0]` and `ax=axes[1]` as this will inform python as to which plot will go in which empty graph

### What is f, axes

f stands for figure  

axes (or ax for short) stands for plot location.  

```python

axes = array([<AxesSubplot:xlabel='BudgetMillions', ylabel='AudienceRating'>,
       <AxesSubplot:xlabel='BudgetMillions', ylabel='CriticRating'>],
      dtype=object)
```
The above result shows that an axes is an array that contains each subplot information

## matrix of subplots

What does matrix in subploting mean?  

If you have **more than one row** then you have matrix of subplots. e.g. `plt.subplots(2,3)`, `plt.subplots(3,3)`,`plt.subplots(3,5)`.

![](subplots.png)

What do you see? 
- more than one row
- you see three rows and three columns, 
- a **table arrangement of plots**

A table is a matrix which is a 2-D object. In order to place the plots in each of the empty plots (cells of matrix), you must specify [row, column]
```python
matrix_name[1,2] 
#This will allow you to access an individual cell present in 2nd row [1] and 3rd column [2]
```
### axes for matrix of subplots  
To place your first plot in the first figure, use ax = axes[0,0]  

To place your 2nd plot in the 2nd figure (same row, but 2nd column), use ax = axes[0,1] and so on..

```python
f,axes = plt.subplots(3,3,figsize=(12,6))
k1 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,ax=axes[0,0])
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,ax=axes[0,1])
```
The result:-  
![](subplot_matrix.png)

## Zooming in the plots - use $xlim=(range)$

```python
f,axes = plt.subplots(3,3,figsize=(12,6))
k1 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,ax=axes[0,0])
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,ax=axes[0,1])
k1.set(xlim=(-20,160))
```

What does (-20,160) mean?  
It is a range. This allows your x-axis to be between -20 and 160.

### To copy zoom setting for all the plots - use $sharex = True$ when specifying the function 

```python
f,axes = plt.subplots(3,3,figsize=(12,6),sharex = True)
k1 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,ax=axes[0,0])
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,ax=axes[0,1])
k1.set(xlim=(-20,160))
```

**Do have same y-axis for all the plots - $sharey = True$**


```python
f,axes = plt.subplots(1,2,figsize=(12,6),sharex = True, sharey= True)
k1 = sns.kdeplot(movies.BudgetMillions,movies.AudienceRating,ax=axes[0])
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating,ax=axes[1])
k1.set(xlim=(-20,160))
```

````python
k2 = sns.kdeplot(movies.BudgetMillions,movies.CriticRating)

#How movie Budget is affecting CriticRating?

#At BudgetMillion of 30:
#The min value is close to -20 and max value is more than 100.

#Conclusion
#The relationship is very similar to what we saw w.r.t. AudienceRating

#difference?
#Change in min and max value is relatively small.
````
