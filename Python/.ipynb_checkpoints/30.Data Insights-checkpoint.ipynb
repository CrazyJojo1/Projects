```python
def myplot(data, playerList=Players):  #I have just added Players as the default value for playerList
    """This function is designed to 
    provide plots for different array"""
    import numpy as np
    import matplotlib.pyplot as plt
    
    %matplotlib inline
    plt.rcParams['figure.figsize']=10,8 
    
    import warnings
    warnings.filterwarnings('ignore')
    
    Cdict = {"KobeBryant":'Blue',"JoeJohnson":'orange',"LeBronJames":'green',"CarmeloAnthony":'red',"DwightHoward":'purple',"ChrisBosh":'brown',"ChrisPaul":'pink',"KevinDurant":'gray',"DerrickRose":'olive',"DwayneWade":'cyan'}
    Markerdict = {"KobeBryant":'o',"JoeJohnson":'v',"LeBronJames":'^',"CarmeloAnthony":'<',"DwightHoward":'>',"ChrisBosh":'s',"ChrisPaul":'p',"KevinDurant":'*',"DerrickRose":'h',"DwayneWade":'d'}
    
    for string in playerList: #playerlist is just like a number generator
        plt.plot(data[Pdict[string]], c=Cdict[string], ls ='dotted', marker=Markerdict[string], ms = 7,label=string)
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.show()
```

```python
def myplot(data, 'title', playerList=Players):  #I have just added Players as the default value for playerList
    """This function is designed to 
    provide plots for different array"""
    import numpy as np
    import matplotlib.pyplot as plt
    
    %matplotlib inline
    plt.rcParams['figure.figsize']=10,8 
    
    import warnings
    warnings.filterwarnings('ignore')
    
    Cdict = {"KobeBryant":'Blue',"JoeJohnson":'orange',"LeBronJames":'green',"CarmeloAnthony":'red',"DwightHoward":'purple',"ChrisBosh":'brown',"ChrisPaul":'pink',"KevinDurant":'gray',"DerrickRose":'olive',"DwayneWade":'cyan'}
    Markerdict = {"KobeBryant":'o',"JoeJohnson":'v',"LeBronJames":'^',"CarmeloAnthony":'<',"DwightHoward":'>',"ChrisBosh":'s',"ChrisPaul":'p',"KevinDurant":'*',"DerrickRose":'h',"DwayneWade":'d'}
    
    for string in playerList: #playerlist is just like a number generator
        plt.plot(data[Pdict[string]], c=Cdict[string], ls ='dotted', marker=Markerdict[string], ms = 7,label=string)
        
    plt.legend(loc='upper left', bbox_to_anchor=(1,1))
    plt.xticks(list(range(10)),Seasons,rotation='vertical')
    plt.title(title) 
    plt.show()
    
    #you need to specify the titles in quotation marks when calling your function 
```

```python
#Again injuries are affecting our plots

#so solution?

#normalize it.

#How?

#compare one in-game matrix with another in-game matrix

myplot(FieldGoals/FieldGoalAttempts, 'FieldGoals/FieldGoalAttempts')   #This gives us accuracy of the player
myplot(MinutesPlayed/Points, 'MinutesPlayed/Points')      

myplot(FieldGoals/Games, 'FieldGoals/Games')
myplot(FieldGoalAttempts/Games, 'FieldGoalAttempts/Games')

myplot(Points/Games, 'Points/Games')
```

```python
#comparing in-game to in-game is better

#How?

#for instance, FieldGoals/Games will give you how many goals they scored in the games they played. 
#This will likely remove the games they couldn't play because of their injury.

```

```python
#Some interesting observations

myplot(MinutesPlayed/Games, "MinutesPlayed/Games") #here injuries does skew our results

#Also, almost all player (except CarmeloAnthony, KevinDurant, maybe few others) are playing less minutesper game since 2005.


myplot(Games, "Games")
#games played is pretty consistent throughout ten years.

#comparing both of these plots.

#What do you derive?

#They might be playing less minutes per game, but they are pretty consistent in playing their games each season.
```

```python
#Since minute played is less and games played is consistent, one can conclude that time is valuable.

#This leads us to efficiency

myplot(FieldGoals/MinutesPlayed, "Efficiency")

#What do you see?

The plot is divided into two section
#5 player above 24% (Kobe, Carmelo, DwayneWade, KevinDurant, LeBron)
#5 players below 24%(JoeJohnson, DerrickRose, DwightHoward, chrisBosh, ChrisPaul)

#Also as per instructor, if you try seeing the plot closely, you will discover one player that is improving consistently
#name?
#KevinDurant.
```
