# Coordinates and diagonals

## Co-ordinates (xlim and ylim)

You can use `xlim` and `ylim` to zoom-in. They are used to limit the co-ordinate axis.

xlim =(0,100) will start x-axis from 0 and end it at 100.

Same is true for y-axis.

### How to set xlim and ylim - `plot_name.set(xlim=(x1,x2),ylim=(y1,y2))`
```python
kws=dict(s=50,linewidth=.5,edgecolor='black')
plot1 = sns.FacetGrid(data=movies,row='Genre',col='Year',hue='Genre')
plot1 = plot1.map(plt.scatter,'CriticRating','AudienceRating',**kws)
#setting the coordinates
plot1.set(xlim=(0,100),ylim=(0,100)) #xlim is to limit the coordinates of x-axis and ylim does the same for y-axis.
```
---

## Diagonal (often a line paritioning the plot in two equal parts)

### To create a line - use  `plot_name.plot((x1,x2),(y1,y2))`

This gives you two points- Point1 = (x1,y1) and Point2 = (x2,y2)
```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
```

### How to convert a 2-D plot to a 1-D plot

A line is 1-Dimensional object, but our plots are 2-Dimensional. 

Plus, we are working with facet grid, which means we have series of plots in rows and columns. To apply $plot()$ to array of plot, we will use using `ForLoop`

But first, we must find a way to apply iteration on each subplot (facet grid plot) - use `.axes`, this will treat each row as one axis.

Since a diagonal is 1-D, we will use `.flat` to treat each array as it is a 1-D array.

```python
for ax in plot1.axes.flat:
    ax.plot((0,100),(0,100),c='orange',ls='--')
```

**diagonal styling-**
- For color, use `c`
- For linestyling, use `ls`

### Benefit of using a diagonal?
Since our x=CriticRating and y=AudienceRating,  
Anything above the line means y-axis is in focus; Audience gave higher rating compared to critic Rating  
Anything below the line means x-axis is in focus; Critic gave higher rating compared to Audience Rating

If still not understood, let us study the graph below-
![](diagonal.png)

Since most points are above the diagonal, between ylim =(20,100), Audience rating which is on y-axis gave more scores than critic Rating.

Each point is (x,y) = (CriticRating, AudienceRating) -   
For instance,   
- (20,50) will put the point above the diagonal. This happened because of AudienceRating (50).

Similary,
- (70,20) will put the point below the diagonal. This happened because of CriticRating (70)

## Complete code will look like this
```python
#facet grid 
Myplot = sns.FacetGrid(data=movies, row='Genre', col='Year',hue='Genre')
#map() - apply function and define x and y labels
Myplot = Myplot.map(plt.scatter,'CriticRating','AudienceRating')

## add co-ordinates - use xlim() and ylim()
Myplot.set(xlim=(0,100),ylim=(0,100))

## add diagonal - use plot()
for ax in Myplot.axes.flat:
    ax.plot((0,100),(0,100),c='gray',ls='dashed')
    
Myplot.add_legend()    
```    



````python
#This lecture is how to tidy up the plots.

#controlling axes and adding diagonals
kws=dict(s=50,linewidth=.5,edgecolor='black')
plot1 = sns.FacetGrid(data=movies,row='Genre',col='Year',hue='Genre')
plot1 = plot1.map(plt.scatter,'CriticRating','AudienceRating',**kws)
#setting the coordinates
plot1.set(xlim=(0,100),ylim=(0,100)) #xlim is to limit the coordinates of x-axis and ylim does the same for y-axis.


#setting up the diagonals
for ax in plot1.axes.flat:
    ax.plot((0,100),(0,100),c='orange',ls='--')
#for this we will be using loops.
#.axes is an array of subplots or the elements of facet grid that we are working with. 
#array here - is list of subplots in single row. (we have 5 plots for each row)

#.axes.flat allows iterating over the array as if it were a 1-D array
#this allows us to treat each row as 1-D array. and then it will apply iteration over it.

#To plot a line (diagonal) - plot((x1,x2),(y1,y2))
#x1,x2 (20,60) and y1,y2 (20,60)- gives you two points - point1(20,20) and point2 (60,60)

#To add style to this line-
#use parameter c - color
#ls (linestyle)

#What does a line represent?
#Since our x=CriticRating and y=AudienceRating,
#Anything above the line means y-axis is in focus; Audience gave higher rating compared to critic Rating
#Anything above the line means x-axis is in focus; Critic gave higher rating compared to Audience Rating

#how to add a legend
plot1.add_legend()
````

````python
Myplot = sns.FacetGrid(data=movies, row='Genre', col='Year',hue='Genre')
#map() - apply function and define x and y labels
Myplot = Myplot.map(plt.scatter,'CriticRating','AudienceRating')

## add co-ordinates
Myplot.set(xlim=(0,100),ylim=(0,100))

## add diagonal - use plot()
for ax in Myplot.axes.flat:
    ax.plot((0,100),(0,100),c='gray',ls='dashed')
````

````python
#Revision

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

plt.show()

sns.set_style('darkgrid')

import numpy as np

import pandas as pd
import os

#importing data

os.getcwd()

movies = pd.read_csv('movies.csv')

#changing column names
movies.columns = ['Film', 'Genre', 'CriticRating', 'AudienceRating',
       'BudgetMillions', 'Year']

#changing data type
movies.Genre = movies.Genre.astype('category')
movies.Film = movies.Film.astype('category')
movies.Year = movies.Year.astype('category')

#check levels
movies.Genre.cat.categories
movies.Genre.unique()

##subsetting and filtering
movies[2:10] #slicing for rows

movies[['CriticRating','AudienceRating']] #similar to [c(a,b,d)]

movies[2:10][['CriticRating','AudienceRating']] #combining both

movies.Genre #selecting entire column

print(movies.Year.cat.categories)


#to get budget for a particular genre

movies[movies.Genre=='Drama'].BudgetMillions

#accessing individual element 
movies.at[2,'Film']
movies.iat[0,0]

sns.lmplot(data=movies,x='AudienceRating',y='CriticRating',fit_reg=False)
````
