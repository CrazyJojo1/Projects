__How to escape Characters__
___

Since formatting in markdown requires the use of very specific characters such as -  
#(hash with space is used for heading),  
*(used for adding italics and un-odered list item),  
**(used to get bold text**,  
+(+ with space is used to create an ordered list item,  
_(used to add italics), etc.  
`(used to add inline code`. 

In order to print these special characters - you would need to `proceed` that specific character with a backslash. e.g.  
\* I am tired  (this will print the asterick symbol without converting it to a list 

\*\* I am adding double astericks in this statement.

\# I am tired 

This is not an \`inline\` code

List of characters that can be escaped using this method -  
```
\ backslash
` backtick
* asterick
_ underscore
{} curly braces
[] square brackets
() parenthesis
# hash
+ plus
- minus
. dot
! exclamation mark
```

__How to add links for topics that exist in the same file?__

Benefit of using it?
- _it is similar to creating a table of content that allows you to reach to a specific topic just by clicking on that link._

Just like you add hyperlinks, you will be using `[name that you want to display](#topic name with each word separated by hypen)`

e.g.
```python
[At a glance](#At-a-glance)

## At a glance

| **Key** | __Value__ |
|:----|------------:|
| Born | 10 July 1856, *Modern-day Croatia* |
| Died | 7 January 1943 (aged 86), *New York City, United States* |
| Resting place | Nikola Tesla Museum, *Belgrade, Serbia* |
| Citizenship | Austrian *(1856-1891)*, American *(1891-1943)* |
| Best Known for | Modern alternating current (AC) electricity supply system |
```

**Note**:  

- These links are case-sensitive. Meaning, At-a-glance has same case letters as 'At a glance'.  
- Don't forget to use hash (#) and hyphen for topics that contain more than one word.






# Jeckyll

Once you have created an .md file, you will need to run Jeckyll.

Jeckyll will create a whole site of individual static html pages

After you get the static pages, you can upload your .md files to any server out there...

## Benefit of using it:- You don't need a data-base

* Faster
Since your server doesn't need to talk to a data base upon page-request, it is a lot faster.  
* Secure
Since you are not talking to any data base, it is lot secure.
