```python
#first how to create an array
import numpy as np
#create a list
mylist = list(range(9))
#print your object list
print(mylist)
#.......................................................np.array() used to convert list to an array and also to a matrix
#convert this list 
myarray = np.array(mylist)
#print your array
print(myarray)

#.............................................................np.arange() used to form array without using list
#how to create an array directly without using a list
myarray2 = np.arange(1,21)
print("myarray2: ",myarray2)

#----------------------------------------------------forming a matrix using np.reshape()
#Now array is 1D object.
#We need to create a matrix, which is 2D
matrix_row_default = np.reshape(myarray2,(5,4))
print("matrix_row_default: ",'\n',matrix_row_default)
#arrange it column wise
matrix_row = np.reshape(myarray2,(5,4),'C')
print("matrix_row: ",'\n',matrix_row)

matrix_column = np.reshape(myarray2,(5,4),order = 'F')
print("matrix_column: ",'\n',matrix_column)

#...................................................parameter of np.reshape()
#'F' is for column-wise arrangement
#'C' is for row-wise arrangement

x = np.arange(1,21)
print("x: ",x )
y = np.arange(11,31)
print("y: ",y)
#np.array(x,y) #This will need a list 

x1 = list(range(1,21))
print("x1: ",x )
y1 = list(range(11,31))
print("y1: ",y)
#------------------------------------forming matrix using np.array() from lists
#create a matrix
matrix_from_list = np.array([x1,y1]) 
matrix_from_list1 = np.array([y1,x1])

print("matrix_from_list: ",'\n',np.array([x1,y1]))
print("matrix_from_list1: ",'\n',np.array([y1,x1]))
#syntax of np.array is np.array([list1, list2, list3 ])

#Accessing individual elements from these matrices
matrix_from_list[1,:]
matrix_from_list[:,7]
matrix_from_list[0,7]

#remember: indexation in python starts from 0 and not from 1.

#.....................................................dictionary

Cdict = {"john":0,"angy":1,"jesus":2}
Cdict
#access values from dictionary using their keys
Cdict["john"]
Cdict["jesus"]

#------------------------------------------------------matrix operation
print("matrix_addition" '\n',matrix_from_list + matrix_from_list1)
print("matrix_subraction" '\n',matrix_from_list - matrix_from_list1)
print("matrix_division" '\n',matrix_from_list / matrix_from_list1)
print("matrix_multiplication" '\n',matrix_from_list * matrix_from_list1)

#......................................................if warning appear due to nan (not a number)
import warnings
warnings.filterwarnings('ignore')

#.......................................................rounding off a matrix

matrix_division = matrix_from_list / matrix_from_list1
np.matrix.round(matrix_from_list / matrix_from_list1,2) #now save this result in an object

matrix_division1 = np.matrix.round(matrix_from_list / matrix_from_list1,2)
print("matrix_division1"'\n',matrix_division1)

#----------------------------------------------------------visualisation using matplotlib
#import two packages when working with statistical modules
import numpy as np
import matplotlib.pyplot as plt

#if you want plot to be in the same notebook 
%matplotlib inline

#if you need to adjust the output of your plot (like height and width of it)
plt.rcParams['figure.figsize'] = 10,8

#let's plot
#plt.plot(matrix_division1)
#..................................plot just the first row with title
plt.plot(matrix_division1[0,:])
plt.title('matrix_division1[0,:]')

plt.show()
#.................................adding aesthetics to the plot
plt.plot(matrix_division1[1,:])
plt.show()
```
