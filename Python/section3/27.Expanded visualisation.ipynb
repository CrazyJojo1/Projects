<u> How to add legend <u/>

Use ```plt.legend()```

How to adjust the location of the legend:-  <br>
    
```python
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
```
   
- loc provides location of legend. e.g.loc ='upper left' means pick the legend from legend's upper left corner. 
-  bbox_to_anchor provides cor-ordinates for graph.  <br>
    e.g. bbox_to_anchor = (1,1) means you want to place the legend at upper right corner of the graph  <br>
Note: You are using tuple (1,1) to give the location of the legend. <br>
   tuple (x,y) is your graph co-ordinates <br>
   (0,0) means the reference point. <br>
   (1,0) means x is of length 1 , but y is at 0. <br>
   (1,1) means both x and y are of length 1 (upper right corner) <br>
   (0,1) gives you upper left corner <br>
