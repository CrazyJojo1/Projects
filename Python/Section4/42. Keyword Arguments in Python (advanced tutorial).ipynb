# Keyword Arguments in Python (kws)

**Keyword arguments (kws) are arguments that you pass to a function (like `.lmplot()`) using dictionaries.**

```python
#linear correlation using lmplot
plot4 = sns.lmplot(data=stats, x="InternetUsers",y='BirthRate',hue='IncomeGroup', fit_reg=False, size = 5, aspect= 1)
```
This plot looks somewhat incomplete - marker size is too small and what about their color... can we change that.

If you press shift tab when using lmplot(), you will see something like this at the very end.
```
{scatter,line}_kws : dictionaries
    Additional keyword arguments to pass to ``plt.scatter`` and
    ``plt.plot``.
```    
What does it mean?  
`lmplot()` of seaborn is using `matplotlib.pyplot.scatter()` to create its visualisation. This means, parameters of matplotlib.pyplot.scatter() can also be used in .lmplot(). 

[To understand more about matplotlib.pyplot.scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)

```python
#In order to increase the size of marker, use scatter_kws = {"s":100}
plot4 = sns.lmplot(data=stats,
                   x="InternetUsers",
                   y='BirthRate',
                   hue='IncomeGroup',
                   hue_order=['Low income','Lower middle income','High income', 'Upper middle income'],                                fit_reg=False, 
                   size = 5, 
                   aspect= 1,
                  scatter_kws={"s":100})
```

## What is {"s":100}?

It is a dictionary that provides keyword argument and its value. `s` is a parameter that defines marker's size.

````python
plot4 = sns.lmplot(data=stats, x="InternetUsers",y='BirthRate',hue='IncomeGroup',
                   hue_order=['Low income','Lower middle income','High income', 'Upper middle income'], fit_reg=False, 
                   size = 5, aspect= 1,
                  scatter_kws={"s":200})
````
