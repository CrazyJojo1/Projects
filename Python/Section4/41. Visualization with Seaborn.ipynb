```python
#Revision

import pandas as pd
import os 

os.getcwd()

stats = pd.read_csv('demographic.csv')
stats

#changing column names
stats.columns =  ['CountryName', 'CountryCode', 'BirthRate', 'InternetUsers',
       'IncomeGroup']

stats

#----------------------------------
import matplotlib.pyplot as plt
import seaborn as sns

print(stats.head())

sns.set(rc={'figure.figsize':(20,10)})

# plot = sns.boxplot(data = stats,x="IncomeGroup", y= "BirthRate")
#sns.set(style="darkgrid")

#kdeplot provide density distribution (much like histogram)
plot2 = sns.kdeplot(stats["InternetUsers"])
plot2.set_title('A distribution chart made with Seaborn')

```

```python
import warnings
warnings.filterwarnings('ignore')
plot3 = sns.distplot(stats["InternetUsers"])

plot3.set_title('This is distplot')
plt.xlabel('InternetUsers', fontweight='bold', color = 'orange', fontsize='17', horizontalalignment='center')
plt.ylabel('Density', fontweight='bold', color = 'orange', fontsize='17', horizontalalignment='center')
```

```python
plot4 = sns.distplot(stats["BirthRate"])
```

```python
#plot internet users w.r.t. Birth rate

#Let's see if we can draw some correlations between the two
vis3 = sns.lmplot(data=stats, x="InternetUsers", y='BirthRate',fit_reg=False, hue="IncomeGroup",col ="IncomeGroup" )

#lm stands for linear model plot (Linear Regression)

#In oder to open the help window - press shift + tab when working with a function

#fit_reg = True (will show you a line in your plot)

#Let's map

#hue is the parameter for color

#col parameter allows you to get different charts for each category

#to control the size of your plot
#size=10
#aspect(width/height)  (by default, aspect =1, width =height) (aspect=2 means width is 2 times the height)
```
