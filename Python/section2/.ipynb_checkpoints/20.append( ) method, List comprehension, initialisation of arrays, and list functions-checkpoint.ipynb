# Using append( ) method to create a list in python

## 1. Add a new element to existing list

- .append() allows you to add a 'single new element' at the end of a list. 

- This increases the length of your list to be increased by 1.

 **Syntax:** <br>
`list(to which you want to add to) . append(elements that you wish to add)`  e.g. sample2.append(sample1)

Below are some of the examples-
```python
sample1 = ["Hello", "sir", 1,2,3,4,8]
sample1.append(10) #i am additional a list of single element to sample1
#see if new element as been added
print(sample1)
sample1 = ['Hello', 'sir', 1, 2, 3, 4, 8, 10] #10 has been added
#check its length
len(sample1) = 8
#-----------------------------------
#append method will only increase the length of a list by 1
sample2 = [50.45,56,78.90,48,3]
sample1.append(sample2)
print(sample1)
sample1 = ['Hello', 'sir', 1, 2, 3, 4, 8, 10, [50.45, 56, 78.9, 48, 3]]
#check the length after using append 
len(sample1) = 9  #note: length is just increased by 1
#------------------------------------
#let's add sample 1 to sample2 
sample2.append(sample1)
print(sample2)
sample2 = [50.45, 56, 78.9, 48, 3, ['Hello', 'sir', 1, 2, 3, 4, 8, 10, [...]]]
```


# List comprehension <br>

 **Syntax : newList = [ expression(element) ```for``` element ```in``` oldList ```if condition``` ]**  <br>
 
<u> - it is short and simple way to create a new list from an existing list when you are working with loops and if statements.</u>  <br>
 Note: you must place the syntax in square bracket; [ ]  <br>
 
 let's see the examples to better understand it-  <br>

## 3. Converting string to float type

If your object is a character type and you need to convert it into numeric data type. Then read below information.

If you have numeric variables listed as string, numbers may look like- `1,003` or `2,345,679`.

In order to convert `1,003` to numeric type, you first need to remove the commas. 

>In R, we used ``gsub()``, but in python, we will make use of list comprehension.

Follow these two steps:- 

**step1- replacing commas from string object.  
step2- Once string is removed. You string will look like `1003`. Use `astype(float64)` to convert it to float type**

```python
Use replace function
s stands for string. 

#step1- replacing commas from string object.
Releases.AdjustedGrossMillions= [s.replace(',','') for s in Releases.AdjustedGrossMillions]

#step2- Once string is removed. You string will look like `1003`. Now you can convert it to float type
Releases.AdjustedGrossMillions= Releases.AdjustedGrossMillions.astype('float64')
```
